```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd()

if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

POWERBI_DATA_DIR = PROJECT_ROOT / "outputs" / "powerbi_data"

fact_sales = pd.read_csv(POWERBI_DATA_DIR / "fact_sales.csv")
dim_date = pd.read_csv(POWERBI_DATA_DIR / "dim_date.csv")
dim_products = pd.read_csv(POWERBI_DATA_DIR / "dim_products.csv")
dim_customers = pd.read_csv(POWERBI_DATA_DIR / "dim_customers.csv")
dim_sellers = pd.read_csv(POWERBI_DATA_DIR / "dim_sellers.csv")
```

```python
tables = {
    "fact_sales": fact_sales,
    "dim_date": dim_date,
    "dim_products": dim_products,
    "dim_customers": dim_customers,
    "dim_sellers": dim_sellers,
}

summary = []

for table_name, df in tables.items():
    summary.append(
        {
            "table_name": table_name,
            "row_count": df.shape[0],
            "column_count": df.shape[1],
            "duplicate_rows": df.duplicated().sum(),
        }
    )

pd.DataFrame(summary)
```
