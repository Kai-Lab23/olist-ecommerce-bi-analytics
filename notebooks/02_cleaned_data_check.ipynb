```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd()

if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

PROCESSED_DATA_DIR = PROJECT_ROOT / "data" / "processed"
```

```python
cleaned_orders = pd.read_csv(PROCESSED_DATA_DIR / "cleaned_orders.csv")
cleaned_order_items = pd.read_csv(PROCESSED_DATA_DIR / "cleaned_order_items.csv")
cleaned_order_payments = pd.read_csv(PROCESSED_DATA_DIR / "cleaned_order_payments.csv")
cleaned_order_reviews = pd.read_csv(PROCESSED_DATA_DIR / "cleaned_order_reviews.csv")
cleaned_customers = pd.read_csv(PROCESSED_DATA_DIR / "cleaned_customers.csv")
cleaned_products = pd.read_csv(PROCESSED_DATA_DIR / "cleaned_products.csv")
cleaned_sellers = pd.read_csv(PROCESSED_DATA_DIR / "cleaned_sellers.csv")

cleaned_orders.head()
```

```python
cleaned_tables = {
    "cleaned_orders": cleaned_orders,
    "cleaned_order_items": cleaned_order_items,
    "cleaned_order_payments": cleaned_order_payments,
    "cleaned_order_reviews": cleaned_order_reviews,
    "cleaned_customers": cleaned_customers,
    "cleaned_products": cleaned_products,
    "cleaned_sellers": cleaned_sellers,
}

summary = []

for table_name, df in cleaned_tables.items():
    summary.append(
        {
            "table_name": table_name,
            "row_count": df.shape[0],
            "column_count": df.shape[1],
            "duplicate_rows": df.duplicated().sum(),
        }
    )

pd.DataFrame(summary)
```
